```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 5, 0.1)   #create x=[0,5] in 0.1 increments
y = np.sin(x)              #y = sin(x)
plt.plot(x,y)              #make a plot
plt.xlabel('x')            #label x axis
plt.ylabel('sin(y)')       #label y axis
plt.show()                 #show the plot on the screen
```
